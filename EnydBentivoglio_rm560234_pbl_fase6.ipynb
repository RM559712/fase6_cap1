```python
#Testar treinamento de v8 em 60 épocas
import os
from ultralytics import YOLO

def get_latest_train_run_folder():
    subfolders = [f.path for f in os.scandir('/content/runs/detect') if f.is_dir() and os.path.basename(f.path).startswith('train')]
    latest_folder = max(subfolders, key=os.path.getctime, default=None)
    return latest_folder

latest_run = get_latest_train_run_folder()

if latest_run:
    weights_path = f'{latest_run}/weights/best.pt'
    if not os.path.exists(weights_path):
        print(f"{weights_path} not found, trying last.pt instead.")
        weights_path = f'{latest_run}/weights/last.pt'

    model = YOLO(weights_path)  # Load best.pt or fallback to last.pt
    results = model.predict(source='/content/drive/MyDrive/FIAP/Teste', imgsz=640, save=True, conf=0.25)
else:
    print("Não foi possível encontrar a pasta de treinamento mais recente.")
```

###Comparação Unificada: YOLOv5 e YOLOv8 (60 e 80 épocas)

---

### Facilidade de Uso/Integração

- *YOLOv5 (60 e 80 épocas)*:
  - Fácil de usar, altamente compatível com Python e Google Colab.
  - Fácil ajuste de configurações e visualização.

- *YOLOv8 (60 e 80 épocas)*:
  - Interface mais moderna e simplificada.
  - Integração direta com Ultralytics Hub

*Conclusão: Ambos são simples, mas o v8 se destaca por ser mais fluido

---

###  Precisão do Modelo (mAP@0.5)

| Modelo           | mAP@0.5 |
|------------------|---------|
| YOLOv5 (60 ep)   | 0.656   |
| YOLOv5 (80 ep)   | 0.630   |
| YOLOv8 (60 ep)   | 0.720   |
| YOLOv8 (80 ep)   | 0.850   |

*Conclusão: Apesar do O YOLOv5 com 60 épocas surpreender positivamente, o YOLOv8 com 80 épocas lidera com folga.

---

### Tempo de Treinamento

| Modelo           | Total estimado |
|------------------|----------------|
| YOLOv5 (60 ep)   | ~3.6 min       |
| YOLOv5 (80 ep)   | ~4.8 min       |
| YOLOv8 (60 ep)   | ~9.0 min       |
| YOLOv8 (80 ep)   | ~6.5 min       |

*Conclusão: O YOLOv5 tem treinamento significativamente mais rápido.

---

### ⚡ Tempo de Inferência

| Modelo           | Tempo (ms/imagem)  |
|------------------|--------------------|
| YOLOv5 (60 ep)   | ~15–25             |
| YOLOv5 (80 ep)   | ~15–25             |
| YOLOv8 (60 ep)   | ~20–35             |
| YOLOv8 (80 ep)   | ~20–35             |

 *Conclusão: Todos são rápidos, mas o YOLOv5 é mais leve.

---

### Conclusão Geral

Para melhores resultados de precisão, o YOLOv8 com 80 épocas é a melhor escolha. Se o objetivo for leveza e velocidade em dispositivos embarcados, o YOLOv5 leva vantagem em qualquer uma das épocas testadas.
